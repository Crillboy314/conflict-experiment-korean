```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Parameters
theta_star = 1
sigma_eta = 1
k = 5
theta_0 = 0

# Define the functions
def A(theta):
    return norm.cdf((theta_star - theta) / sigma_eta)

def g(theta):
    return 1 / (1 + np.exp(-k * (theta - theta_0)))

# Define theta range
theta = np.linspace(-3, 3, 400)

# Compute the functions
A_theta = A(theta)
g_theta = g(theta)

# Plotting the functions
plt.figure(figsize=(10, 6))

plt.plot(theta, A_theta, label='$A(\\theta)$', color='blue')
plt.plot(theta, g_theta, label='$g(\\theta)$', color='red')

plt.axhline(y=0, color='gray', linestyle='--')
plt.axhline(y=1, color='gray', linestyle='--')
plt.xlabel('$\\theta$')
plt.ylabel('Function value')
plt.title('Example functions $A(\\theta)$ and $g(\\theta)$')
plt.legend()
plt.grid(True)
plt.show()
```
